```python
import networkx as nx
import scipy.stats as sp
import numpy as np
```

## Centrality functions

```python
g = nx.generators.small.krackhardt_kite_graph()
```

```python
nx.draw(g)
```

```python
degree_cen = nx.degree_centrality(g)
degree_cen
```

```python
between_cen = nx.betweenness_centrality(g)
between_cen
```

```python
close_cen = nx.closeness_centrality(g)
close_cen
```

```python
eigen_cen = nx.eigenvector_centrality(g)
eigen_cen
```

## Correlation coefficients

```python
### let's take two random vectors
x = np.arange(10,20)
y = np.array([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])

pearson = sp.pearsonr(x,y)[0]
print(f"Pearson's correlation: {pearson}")

kendall = sp.kendalltau(x,y)[0]
print(f"Kendall's tau: {kendall}")

spearman = sp.spearmanr(x,y)[0]
print(f"Spearman's rho: {spearman}")
```

## PageRank

```python
g = nx.generators.directed.gn_graph(20, seed=1234)
nx.draw(g)
```

```python
pr85 = nx.pagerank(g, alpha=0.85)
print(pr85)
pr35 = nx.pagerank(g, alpha=0.35)
print(pr35)
```

```python
import matplotlib.pyplot as plt
plt.scatter([pr85[x] for x in pr85], [pr35[y] for y in pr35])
plt.xlabel("pr(0.85)")
plt.ylabel("pr(0.35)")
```
