```python
import networkx as nx
```

```python
g = nx.generators.small.krackhardt_kite_graph()
```

```python
nx.draw(g, with_labels = True)
```

```python
# List comprehension
nodes = [i for i in g.nodes() ]
```

```python
nodes
```

```python
# 'degrees' list is alligned with 'nodes' list
degrees = [g.degree(i) for i in nodes]
```

```python
degrees
```

```python
nx.clustering(g)
```

```python
nx.average_clustering(g)
```

```python
nx.average_degree_connectivity(g)
```

```python
nx.degree_assortativity_coefficient(g)
```

```python
nx.draw(g)
```

```python
grnd = nx.algorithms.smallworld.random_reference(g,niter = 10, connectivity=False)
```

```python
nx.draw(grnd)
```

```python
nx.degree_assortativity_coefficient(g)
```

```python
nx.degree_assortativity_coefficient(grnd)
```
